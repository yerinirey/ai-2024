### K-평균 알고리즘  
1. 무작위로 k개의 클러스터 중심 지정
2. 각 샘플에서 가장 가까운 중심을 찾아 해달 클러스터의 샘플로 지정
3. 샘플의 평균값으로 클러스터 중심을 각각 변경
4. 중심에 변화가 없을 때까지 2-3 반복

```python
!wget https://bit.ly/fruits_300 -O fruits_300l.npy
```

```python
import numpy as np

fruits = np.load('fruits_300l.npy') # (샘플 수, 높이, 너비)
fruits_2d = fruits.reshape(-1, 100*100) # (샘플 수, 높이 * 너비)
```

```python
from sklearn.cluster import KMeans
# n_clusters: 클러스터 개수 지정
km = KMeans(n_clusters = 3, random_state = 42)
km.fit(fruits_2d)
```

```python
# 군집의 결과는 labels_ 속성에 배열 형태로 저장된다.
# n_clusters=3인 경우, 0, 1, 2 중 하나의 값을 가짐
print(km.labels_)
```

```python
# 클래스(군집) 분포 확인
print(np.unique(km.labels_, return_counts=True)) # return_counts가 합계를 반환해줌
```

#### 클러스터 내 사진 시각화

```python
import matplotlib.pyplot as plt

def draw_fruits(arr, ratio = 1):
    n = len(arr) # sample개수

    rows = int(np.ceil(n / 10)) # 한 줄에 10개씩 그릴 것. -> 행 = n/10개
    cols = n if rows < 2 else 10
    fig, axs = plt.subplots(rows, cols,
                            figsize=(cols * ratio, rows * ratio), squeeze=False)

    for i in range(rows):
        for j in range(cols):
            if i * 10 + j < n: # n개 까지 그리기
                axs[i, j].imshow(arr[i * 10 + j], cmap='gray_r')
            axs[i,j].axis('off')
    plt.show()
```

```python
draw_fruits(fruits[km.labels_ == 0])
```

```python
draw_fruits(fruits[km.labels_ == 1])
```

```python
draw_fruits(fruits[km.labels_ == 2])
```

#### 클러스터 중심 시각화

```python
# cluster_centers_ 속성에 저장된 클ㄹ스터 중심값 그리기
draw_fruits(km.cluster_centers_.reshape(-1, 100, 100), ratio = 3)
```

#### 각 클러스터 중심값까지 거리 계산  

```python
# transform() 샘플로부터 클러스터 중심까지 거리로 변환하는 메소드
print(km.transform(fruits_2d[100:101]))
```

```python
# 위의 결과에서 0번 클러스터의 거리가 가장 짧음, predict로 클러스터 예측
print(km.predict(fruits_2d[100:101]))
```

```python
# 0번은 파인애플일 것으로 예측된다(cluster_centers_)
# 샘플을 직접 확인
draw_fruits(fruits[100:101])
```

### 최적의 K 찾기 - 엘보우 방법

k평균 알고리즘은 **사전에 k를 지정해야 하고 어떤 k가 적절한지** 알 수 없다.  
- __엘보우 방법__  
  - **k를 바꿔가며** KMeans 실행
  - 각 **클러스터의 샘플끼리 가까우면** 좋은 클러스터
  - 이너셔(inertia) 값을 계산해 샘플들이 얼마나 가까운지 알 수 있음, 값이 낮을수록 가까움
  - 일반적으로 **k가 커지면 inertia가 작아진다**
  - k가 증가시 이너셔 감소 정도가 꺾이는 부분 발생 -> 최적지점

```python
inertia = []
for k in range(2, 7): # k값을 2부터 6까지 증가시킴
    km= KMeans(n_clusters = k, random_state = 42) # k를 바꿔가며 KMeans 실행
    km.fit(fruits_2d)
    inertia.append(km.inertia_)

plt.plot(range(2, 7), inertia)
plt.show()
```
