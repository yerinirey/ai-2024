**차원 축소 알고리즘(PCA(주성분 분석)) 모델 구축**

특성 수가 2개인 경우 평면상에 시각화가 가능하지만, n개일 때는 나타낼 수 없다.  

과일 사진은 10000개의 pixel(특성)으로 구성된 데이터다.  
차원을 줄일 수 있다면 **저장공간 최적화, 시각화** 용이  
-> ***차원축소*** 알고리즘 ( 비지도학습 )

### 주성분 분석 ( Principal Component Analysis )  
1. 데이터에 있는 *분산이 큰 방향*을 찾는 방법  
    방향성을 띠므로 *벡터*로 나타낼 수 있다
2. 1번 주성분에 직교하는 벡터

최대로 지닐 수 있는 주성분은 특성 수와 같다.

#### PCA scikit-learn 구현

```python
import numpy as np

fruits = np.load('fruits_300l.npy')
fruits_2d = fruits.reshape(-1, 100 * 100)
```

```python
# 과일 사진 특성(100*100)을 대상으로 50개를 주성분으로 지정
from sklearn.decomposition import PCA

pca = PCA(n_components=50)
pca.fit(fruits_2d)
```

```python
# components_ 속성에 주성분 저장
print(pca.components_.shape) # (주성분 수, 원본 특성 수)
```

```python
# 각 주성분을 원본처럼 100 * 100 이미지로 출력해보기
import matplotlib.pyplot as plt

def draw_fruits(arr, ratio = 1):
    n = len(arr) # sample개수

    rows = int(np.ceil(n / 10)) # 한 줄에 10개씩 그릴 것. -> 행 = n/10개
    cols = n if rows < 2 else 10
    fig, axs = plt.subplots(rows, cols,
                            figsize=(cols * ratio, rows * ratio), squeeze=False)

    for i in range(rows):
        for j in range(cols):
            if i * 10 + j < n: # n개 까지 그리기
                axs[i, j].imshow(arr[i * 10 + j], cmap='gray_r')
            axs[i,j].axis('off')
    plt.show()
```

```python
draw_fruits(pca.components_.reshape(-1, 100,100))
# 원본 데이터에서 분산이 큰 방향을 순서대로 나타낸 것이다. 샘플이 아님
```

### 주성분 투영 ( 차원 축소 )  
원본 특성(10000차원) -> 주성분(50차원) 으로 투영 == 차원 축소의 효과  

```python
print(fruits_2d.shape) # (원본 샘플 수, 원본 특성 수)
```

```python
fruits_pca = pca.transform(fruits_2d)
```

```python
print(fruits_pca.shape) # (원본 샘플 수, 주성분 수)로 변환된 것을 확인 가능
```

### 원본 데이터 재구성  
축소된 차원(50개의 주성분) 으로부터 10000개의 원본 특성으로 *재구성* 가능  

데이터 손실은 불가피하지만 **선택된 50개 성분이 분산이 큰 방향** 으로 **많은 정보를 담고 있다**.   
때문에 어느정도 복구가 가능하다

```python
fruits_inverse = pca.inverse_transform(fruits_pca)
print(fruits_inverse.shape) # 50 -> 10000
```

```python
fruits_reconstruct = fruits_inverse.reshape(-1, 100, 100) # 100*100
```

```python
for start in [0, 100, 200]:
    draw_fruits(fruits_reconstruct[start:start+100])
    print('\n')
```

### 설명된 분산  
주성분이 원본 data의 분산을 얼마나 잘 나타내는지 나타냄

```python
print(np.sum(pca.explained_variance_ratio_))
```

```python
plt.plot(pca.explained_variance_ratio_)
```

```python
# PCA의 n_components에 원하는 "설명된 분산 비율"을 입력하여 주성분 개수 선택 가능

pca = PCA(n_components = 0.5)
pca.fit(fruits_2d)
```

```python
print(pca.n_components_) # 주성분
```

```python
fruits_pca = pca.transform(fruits_2d)
print(fruits_pca.shape)
```

```python
# 주성분 2개에 대해 k-평균 군집화 수행
from sklearn.cluster import KMeans

km = KMeans(n_clusters=3, random_state = 42)
km.fit(fruits_pca)
```

```python
print(np.unique(km.labels_, return_counts=True))
```

```python
draw_fruits(fruits[km.labels_==0])
```

```python
draw_fruits(fruits[km.labels_==1])
```

```python
draw_fruits(fruits[km.labels_==2])
```

오류가 있기는 하나, 두 개 특성만으로 좋은 군집을 찾음

```python
# 주성분 분석으로 차원 감소 시 시각화에도 용이
for label in range(0, 3):
    data = fruits_pca[km.labels_ == label]
    plt.scatter(data[:,0], data[:, 1]) # 두 개의 주성분을 x,y축으로 삼아 산점도 표시

plt.legend(['apple','banana','pineapple'])
plt.gca().invert_xaxis()
plt.show()
```
